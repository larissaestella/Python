```python
# Validação de inteiro usando função

def faixa_int(minimo,maximo):
    while True: #loop infinito 
        v = int(input("Entre com um valor entre {} e {}:".format(minimo, maximo)))
        if v < minimo or v > maximo:
            print("Valor inválido. Digite um valor entre {} e {}".format(minimo, maximo))
        else: 
            return v
        
minimo = 5
maximo = 10

valor = faixa_int(minimo, maximo)
print("Valor válido:", valor)
```

```python
# Ex. 1 - Validação de inteiro usando função

def palavra(min, max):
    while True:
        s = str(input("Digite uma palavra com o caracteres entre {} e {}:". format(min, max)))
        tam_str = len(s)
        if tam_str < min or tam_str > max:
            print("Palavra inválida. Digite uma nova")
        else:
            return s
        
min = 4
max = 9

print("Palavra válida:", palavra(min, max))
```

```python
def validar_string(string, minimo, maximo):
    tamanho = len(string)
    return minimo <= tamanho <= maximo

resultado = validar_string("Letra", 1, 5)
print(resultado)
```

```python
# Ex.2 - Validação de inteiro usando função

def verificar_string(string, lista):
    return string in lista

nomes = ['maria', 'jose', 'joão', 'ana']
resultado = verificar_string('jose', nomes)
print(resultado)
```

```python
# Exercicío sobre empacotamento e desempacotamento de parâmetros

def soma(a, b):
    return a + b
def sub(a, b):
    return a - b
def produto(a, b):
    return a * b
def div(a, b):
    return a / b

L_soma = [2, 3]
L_sub = [4, 2]
L_produto = [3, 5]
L_div = [20, 4]

resultado_soma = soma(*L_soma)
resultado_sub = sub(*L_sub)
resultado_produto = produto(*L_produto)
resultado_div = div(*L_div)

def media(resultado_soma, resultado_sub, resultado_produto, resultado_div):
    resultado = ((resultado_soma + resultado_sub + resultado_produto + resultado_div) / 4)
    return(resultado)
```
