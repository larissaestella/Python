```python
itens = ['milho', 'feijão', 'leite', 'açúcar', 'arroz', 'azeitona', 'café em grãos']
estoque = [20, 50, 15, 30, 5, 30, 40]
preços = [5, 5.0, 2.0, 6.0, 10, 3.0, 3.0]

for i in range(len(itens)):
    total = estoque[i]* preços[i]
    if total < 120:
        print(f"Produtos: {itens[i]}, Estoque: {estoque[i]}, Preço: R${preços[i]:.2f}, Total: R${total:.2f}")
```

```python
data = {'Nome': ['Alice', 'Bob', 'Charlie', 'Ana', 'Alice', 'Bob'], 
        'Idade': [25, 30, 22, 22, 38, 35], 
        'País': ['Brasil', 'Inglaterra', 'Estados Unidos', 'Itália', 'Austrália', 'México'],
}
covid = []
indice = 0

df = pd.DataFrame(data)

while indice < len(df):
    pais = df['País'][indice]
    if pais == 'Brasil' or pais== 'Itália':
        covid.append("Tem Covid!")
    else:
        covid.append("Não tem Covid")
    indice += 1
    

df['Status'] = covid

df
```

```python
vacinas = []
prioridade = []

indice = 0

while indice < len(df):
    vacina = df['Status'][indice]
    if vacina == "Tem Covid!":
        vacinas.append('Precisa descansar')
    else:
        vacinas.append('Precisa tomar vacina')

    idade = df['Idade'][indice]
    if idade > 35:
        prioridade.append('Alta')
    else:
        prioridade.append('Baixa')
    indice += 1

df['Vacina'] = vacinas
df['Prioridade'] = prioridade
df
    
```
